```python
%matplotlib inline

from __future__ import division
import pandas as pd
import numpy as np
from seaborn import plt
from sklearn.datasets import load_iris
from sklearn import preprocessing
from sklearn.preprocessing import scale
```

```python
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets, feature_selection
from sklearn.cross_validation import train_test_split, cross_val_score
```

```python
%load_ext sql
```

```python
wine = wine_test = pd.read_csv('../data/wine.csv')
wine_test = wine_test.drop(['Wine'], 1)
wine
```

```python
# Create the training (and test) set using scikit-learn's train_test_split function
X_train, X_test, y_train, y_test = train_test_split(wine_test, wine['Wine'], test_size=0.3, random_state=42)
```

```python
wine[['Wine', 'Proline']].groupby('Wine').count() # looks like 3 unique groups?
```

```python
# Various variables we'll need to set intially.
n_neighbors = range(1, 51, 2)
```

```python
# Loop through each neighbors value from 1 to 51 and append
# the scores
scores = []
for n in n_neighbors:
    clf = neighbors.KNeighborsClassifier(n)
    clf.fit(X_train, y_train)
    scores.append(clf.score(X_test, y_test))
```

```python
plt.figure(figsize=(18,6))
_ = plt.plot(n_neighbors, scores, linewidth=3.0)
```

```python
from sklearn.cross_validation import cross_val_score
clf = neighbors.KNeighborsClassifier(29, weights='uniform')
scores = cross_val_score(clf, wine.values, wine['Wine'], cv=5)
print scores
```

```python
np.mean(scores)
```

```python
# Below returns highest signifiance for features 2 and 3
# (remember, Python uses index 0). 
n = np.arange(len(wine_test.columns))

fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(111)

ax.bar(n, feature_selection.f_classif(wine_test, wine['Wine'])[0])

xtickNames = ax.set_xticklabels(wine_test.columns)
ax.set_xticks(n)
_ = plt.setp(xtickNames,rotation = 30, fontsize=12)  # any idea why my plots labels are all screwed up?
```

```python
wine_test.columns
```

```python
clf = neighbors.KNeighborsClassifier(30, weights='uniform')
clf.fit(wine_test[['Flavanoids', 'Proline']].values, wine['Wine'])
```

```python
h = .5  # step size in the mesh
# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
```

```python
# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, m_max]x[y_min, y_max].
x_min, y_min = wine_test.min()[['Flavanoids', 'Proline']]
x_max, y_max = wine_test.max()[['Flavanoids', 'Proline']]
```

```python
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
# Put the result into a color plot
Z = Z.reshape(xx.shape)
```

```python
plt.figure(figsize=(18,6))
plt.pcolormesh(xx, yy, Z, cmap=cmap_light) # also tried with other variables and seemed to have more shape in the boundaries but this seems just "ok"..
```

```python
# Plot also the training points
plt.figure(figsize=(18,6))
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
plt.scatter(wine_test['Flavanoids'], wine_test['Proline'], c=wine['Wine'], cmap=cmap_bold)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title("3-Class classification (k = {}, weights = '{}')".format(clf.n_neighbors, clf.weights))
```

## K-Means / Heir

```python
from sklearn.cluster import KMeans
from seaborn import plt, color_palette
```

```python
kmeans             =  KMeans(3, random_state=1)
wine['k_predict']  =  kmeans.fit(wine).labels_

c_hat = [color_palette()[i] for i in wine['k_predict']]

mu = kmeans.cluster_centers_

plt.scatter(mu[:,list(wine.columns).index('Flavanoids')], mu[:,list(wine.columns).index('Proline')], c=c_hat, s=75)
plt.scatter(wine['Flavanoids'], wine['Proline'], c=c_hat, alpha=0.8)

# Any way I can get these cluster centers to be the same color?
```

```python
# shift groups index +1 to match "Wine" categorical
wine.k_predict = wine.k_predict.apply(lambda x: x+1)
wine
```

# K-Means counts vs Original

```python
wine.groupby('k_predict').count().ix[:,-1]
```

```python
wine.groupby('Wine').count().ix[:,-1]
```

```python
wine.k_pass = wine.Wine == wine.k_predict
wine.k_pass.value_counts()
```

```python
# tell me there's a function that does this in pandas?
for i in range(1,4):
    print 'Wine %d accuracy:' % i, wine['Wine'].loc[wine['Wine'] == i][wine['k_pass'] == True].count() / wine['Wine'].loc[wine['Wine'] == i].count()

```

# Heir / Dendrogram

```python
# compute distance matrix
from scipy.spatial.distance import pdist, squareform

# not printed as pretty, but the values are correct
distx = squareform(pdist(wine, metric='euclidean'))
distx

# perform clustering and plot the dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram

R = dendrogram(linkage(distx, method='single'), color_threshold=100) # color_threshold=100

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram', fontweight='bold', fontsize=14);
```

```python
from sklearn.cluster import AgglomerativeClustering
hierarch = AgglomerativeClustering(n_clusters=3, linkage='ward')

wine['Y_hat_hier'] = hierarch.fit(wine).labels_
c_hat = [color_palette()[i] for i in wine['Y_hat_hier']]
plt.scatter(wine['Flavanoids'], wine['Proline'], c=c_hat);
```
